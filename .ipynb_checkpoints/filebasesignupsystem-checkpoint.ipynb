```python
import smtplib
import random
from twilio.rest import Client
import time

def signup():
    name=input("enter your name")
    email=input("enter your email")
    try:
        fo=open(email+".txt",'r')
        fo.close()
        notexist=0
        print("user already exist")
    except FileNotFoundError:
        notexist=1
    if notexist==1:
        password=input("enter your password")
        city=input("enter your city")
        phone=input("enter your phone number")

        otp=str(random.randint(1000,9999))
        try:
            s=smtplib.SMTP('smtp.gmail.com',587)
            s.starttls() 
            s.login('itronixml@gmail.com','your password')
            msg="This is verification "+otp
            s.sendmail('itronixml@gmail.com',email,msg)
            s.quit()
            success=1
        except :
            print("error in sending email")
            success=0
        if success==1:
            enterotp=input("enter 4 digit otp")
            if(otp==enterotp):
                fo=open(email+".txt",'w')
                writeinfile=name+"\n"+email +"\n"+ password + "\n" +city+ "\n" +phone
                fo.write(writeinfile)
                fo.close()
    
        
    
def signin():
    loginemail=input("enter email")
    loginpass=input("enter password")
               
    try:
        fo=open(loginemail+".txt",'r')  
        lines=fo.readlines()
        passs=lines[2]
        fo.close()
        if(passs==(loginpass+"\n")):
            fo=open(loginemail+".txt",'r') 
            data=fo.read()
            print("your credentials are  ")
            print(data)
            fo.close()
        else:
            print("wrong password")
        
    except FileNotFoundError:
        print("Kindly sign up first")


def forgetpass():
    emailforget=input("enter your registered email")
                 
    try:
        
        otpfor=str(random.randint(1000,9999))
        fo=open(emailforget+".txt",'r')
        lines1=fo.readlines()
        phoneveri=lines1[4]
        fo.close()
        account_sid = 'AXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX'
        auth_token = 'your taken id'
        client = Client(account_sid, auth_token)

        message = client.messages.create(
        body='this is your otp for reseting your password?'+otpfor,
        from_='your teillio number',
        to="+91"+phoneveri )
        print(message.sid)
        time.sleep(5)
        
        print("otp has been sent to "+phoneveri)
        
        otpconf=input("enter otp to change password")
        if(otpfor==otpconf):
            print("you can change password")
            fo=open(emailforget+".txt",'r')
            lines1=fo.readlines()
            fo.close()
            newpassword=input("enter new password")
            lines1[2]=newpassword+"\n"
            fo=open(emailforget+".txt",'w')
            fo.writelines(lines1)
            fo.close()
            print("password changed succesfully")
        else:
            print("wrong otp")
        
        
        
    except FileNotFoundError:
        print("your email is not registerd with us")
    

choice=input("Press Y to start application")
if(choice=='Y'):
    start=1
    while(start):
        print("Enter 1 to sign up")
        print("Enter 2 to sign in")
        print("Enter 3 to forget password")
        print("Enter 4 to to exit")
        option=int(input("enter your choice"))
        if(option==1):
            signup()
        elif(option==2):
            signin()
            
        elif(option==3):
            forgetpass()
        else:
            start=0
            
        
    
```
